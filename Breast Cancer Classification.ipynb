#model Training (finding our solution)

```python
#here we droping our last column i.e target with drop method  and in all column show the axis is 1
x = df_cancer.drop(['target'],axis =1)
```

```python
x
```

```python
#define our output that is target
y=df_cancer['target']
```

```python
y
```

```python
#we want to split our data for training and testing the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
#to show how it work press shift+tab
```

```python
x_train,X_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state= 5)
```

```python
x_train
```

```python
y_train
```

```python
x_train
```

```python
y_test
```

```python
#Here we use support vector machine(svm) to do the classification
```

```python
from sklearn.svm import SVC
```

```python
#here we add classification report and confusion matrix
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# here we instanciate an object
```

```python
svc_model = SVC()
```

```python
#In order to perform traing we use  fit method to train our model
```

```python
svc_model.fit(x_train,y_train)
```

STEP#5

#Evaluating The Model

```python
#here we predict for new model that never trined

y_predict = svc_model.predict(X_test)
```

```python
#here it showing all 1's then it mean the model is not trined
y_predict
```

```python
#here we use confussion matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm)  #it is without annotation
```

```python
sns.heatmap(cm,annot =True)  # it shows annotation
```

# Improving Our Model

```python
#1.Data Normalisation
#2SVM parameters Optimization
  # c-parameter- it can classify the penulty
   #Small C(loose)--
   #Large(strict)--overfitting
```

```python
#Gamma parameter :control how faer the influence of a single training set reaches
   # Large gamma--close reach(close data points have high weight)
  #Small gamma: far reach (more generalized solution)
```

# #6 Improving the model Part-1

```python
min_train = x_train.min()
```

```python
range_train = (x_train - min_train).max()
```

```python
x_train_scaled =(x_train - min_train)/range_train
```

```python
#here we use scatterplot without sacling
sns.scatterplot(x = x_train['mean area'], y = x_train['mean smoothness'], hue = y_train)
```

```python
#here we use sctter plot with scaling
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue = y_train)
```

#here we use same on Testing set

```python
min_test =X_test.min()
range_test =(X_test - min_test).max()
X_test_scaled =(X_test - min_test)/range_test
```

```python
#here we train the model again
svc_model.fit(x_train_scaled,y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
#Great result
#here 43 and  66 correct classification and only 5 miss classified 
```

```python
#classification Report
```

```python
print(classification_report(y_test, y_predict))
```

# Improving The Moodel -PART-2

```python
param_grid = {'C': [0.1,1,10,100],'gamma': [1,0.1,0.01,0.001],'kernel': ['rbf']}
```

```python
#Importing Grid search cv from sk learn
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit = True, verbose=4)
```

```python
grid.fit(x_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
#here we directly run without running previous
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,grid_predictions)
```

```python
sns.heatmap(cm,annot=True)
```

```python
#It's better than previous and improved from previous only 3 miss classified 
```

```python
#Here we print the  classification report
print(classification_report(y_test,grid_predictions ))
```
